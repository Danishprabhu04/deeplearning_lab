```python
!pip install gensim numpy

# Cell 2: Import libraries
import gensim
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.test.utils import common_texts
import numpy as np
```

```python
# Cell 3: Train Word2Vec and load GloVe
# Train Word2Vec Model
model_w2v = Word2Vec(sentences=common_texts, vector_size=50, window=5, min_count=1, workers=4)
model_w2v.save("word2vec.model")

# Load pretrained GloVe embeddings (download glove.6B.50d.txt from https://nlp.stanford.edu/projects/glove/ if needed)
embedding_index = {}
try:
    with open("glove.6B.50d.txt", encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            embedding_index[word] = np.array(values[1:], dtype=np.float32)
    print("GloVe embeddings loaded successfully.")
except FileNotFoundError:
    print("GloVe file not found. Please download glove.6B.50d.txt and place it in the current directory.")
    embedding_index = {}

# Print most similar words for 'computer' from Word2Vec model
try:
    print("Word2Vec most similar to 'computer':", model_w2v.wv.most_similar('computer'))
except KeyError:
    print("'computer' not in Word2Vec vocabulary.")

# Optional: Print some GloVe embeddings
if embedding_index:
    glove_embedding = embedding_index.get('computer', 'Not found')
    print("GloVe embedding for 'computer':", glove_embedding)
else:
    print("GloVe embeddings not loaded.")
```
